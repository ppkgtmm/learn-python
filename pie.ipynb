### control flow
- elif (else if), else are optional and both should come after an if

```python
pets = ["fish"]

if "dog" in pets:
	# start block (indentation)
	print("you have a dog.")
	# end block
elif "fish" in pets: # evaluates to True (condition met)
  print("you have a fish") # executed
else:
  print("not sure!") # executed if no conditions above is satisfied
```

### loops
#### while loops
- repeatedly execute code block as long as a condition evaluates to `True`.
- condition of a `while` loop is always checked first before the block of code runs.

```python
i = 1
while i < 6:
  print(i)
  i = i + 1
```

should not iterate and modify iterable at the same time, instead create new iterable to store the results get from changing

### continue
continues with the next iteration of the loop

```python
for w in words:
  if len(w) < 5:
    continue
  print(w, len(w))
```

### break
breaks out of the innermost loop

```python
i = 1
flag = False
while not flag:
  while True:
    if i == 6:
      break
    print(i)
    i = i + 1
  flag = True
```

### try except finally
- used to handle error. code which may raise an error is written in the `try` block.
- if the code raises an error, the rest of the `try` block will cease executing and the `except` code block will execute.
- code in `finally` block will be executed when either `try` or `except` block has finished execution

```python
try:
    raise Exception # throw exception raise exception_name
except:
    print("error")
finally:
    print("finally")
```

### default arguments
used to provide default values to function when argument is not passed <br>
- default arg is shared across multiple calls
- if default args will be mutated `( mutation != reassignment)` and we don't want to share them accross calls use `None` as default value and later initialize the default arg in function
- when defining func argument with no default must not follows argument with default value

```python
def f(a, L=None):
    if not L: 
        L = [] # reassign if no list provided
    L.append(a)
    return L

'''
below used positional argument where python determine mapping 
of param name and value by position
'''
f(1), f(2), f(3, [])
```

```python
def g(x=0, y):
  return x + y
```

### keyword argument
- pass `arg_name = value` when calling func to help python determine mapping of param name and its value

- positional argument must not follow keyword argument

```python
f(a=2), f(L=[], a=2), f(2, L=[])
```

```python
f(a=0, [])
```

```python
f(0, L=[], a=0) # if extra argument(s) is/are passed to python func, error is thrown
```

### arguments
- standard arguments can be pass to function as either
positional arg or keyword arg <br>
#### for python 3.9+
- to define how we can pass arguments to functions, we can use `/` and `*` parameters as separators when defining function
- use positional args when not want to expose parameter names to user or when want to enforce the order of the args on func call.
- use keyword args when the func is more understandable by exposing arg names or if want to be more flexible for args order




```python
'''
pos arg(s), / , standard arg(s), * , keyword arg(s)
when calling function we don't need to pass  `/` and `*` 
'''
def combined_example(pos_only, /, standard, *, kwd_only):
  print(pos_only, standard, kwd_only) 
```

### unfixed arg length

```python
def concat(*args, sep="/"):
  print('args info:', type(args), args, *args)
  return sep.join(args)

print(concat(), concat('a', 'b'))
```
