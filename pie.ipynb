### try except finally
- used to handle error. code which may raise an error is written in the `try` block.
- if the code raises an error, the rest of the `try` block will cease executing and the `except` code block will execute.
- code in `finally` block will be executed when either `try` or `except` block has finished execution

### default arguments
used to provide default values to function when argument is not passed <br>
- default arg is shared across multiple calls
- if default args is mutable `( mutation != reassignment)` and we don't want to share them accross calls use `None` as default value and later initialize the default arg in function
- when defining func argument with no default must not follows argument with default value

```python
def f(a, L=None):
    if not L: 
        L = [] # reassign if no list provided
    L.append(a)
    return L

'''
below used positional argument where python determine mapping 
of param name and value by position
'''
f(1), f(2), f(3, [])
```

```python
def g(x=0, y):
  return x + y
```

### format string
format strings or f strings help include the value of python expressions inside a string <br>

```python
table = {1: ('Sjoerd', 56.79), 100: ('Jack',78.1025), 1001: ('Dcab', 79.002369)}
# format table headers, use provided python string 'id', 'name' etc. and space for each header string
print(f"""{'id':4} | {'name':10} | {'weight':7}""")
for id, data in table.items():
  '''
  format table columns values with provided space
  make 4 digits space for integer id (:4d)
  make 10 char space for string data[0] (:10) and enclose the string with quotes (!r)
  make 7 digits space including decimal points for float data[1] (:7) (not enough space for 79.002369)
  if we didn't specify enough space python will auto adjust spaces for us without making data loss
  '''
  print(f'{id:4d} | {data[0]!r:10} | {data[1]:7f}') 

# format mean weight value from computation to contain 2 decimal points (.2f)
print(f"""{'mean weight':17} | {sum([w for n, w in table.values()])/len(table):.2f}""")
```

read from a file
- if file not exist throw error
- if file exist read data from it



```python
def read_file(file_name):
  with open(file_name, 'r') as f:
    r = f.read() # read all data at once
    return r

print(read_file('workfile'))
```

### JSON read write

use when want to save more complex data types with less effort

- json can take python data hierarchies, and convert them to string representations; this process is called serializing.
- reconstructing the data from the string representation is called deserializing.

#### write
create new file when file not exist and write



```python
import json
with open('json_json', 'w') as f:
  json.dump((1,2,3), f) # serialize and save to file as json string

f.closed
```

#### read
throw error when file not exist

```python
with open('json_json', 'r') as f:
  r = json.load(f) # deserialize and store parsed data to variable

print(r, tuple(r))
f.closed
```
