### set
contains unique items

```python
fruits = {"apple", "banana", "cherry"}
```

### set methods

```python
fruits.add("orange") # adds an element to the set (mutate set)
fruits.add("apple") # item not added as alraedy exist
print(fruits)
```

```python
print(fruits.copy()) # return a copy of the set
print(fruits)
```

```python
fruits.difference({'apple'}) # return all items except items in set passed as arg
```

```python
fruits.intersection({'banana', 'cherry', 'kiwi'}) # return items exist in both sets
```

```python
# returns items that are present in both sets at the same time
fruits.symmetric_difference({'banana', 'cherry', 'blueberry'})
```

```python
# merge set with other iterable(s)
print(fruits.union({'cherry', 'blueberry'}, ['orange', 'kiwi'], ('rambutan',)))
print(fruits)
```

```python
# same as union (mutate set)
fruits.update({'cherry', 'blueberry'}, ['orange', 'kiwi'], ('rambutan',))
print(fruits)
```

```python
fruits.remove('banana')
fruits.discard('rambutan')
print(fruits)
```

```python
fruits.discard('rambutan') # not throw error if item not exist in set
fruits.remove('coconut')
```

### control flow
- elif (else if), else are optional and both should come after an if

```python
pets = ["fish"]

if "dog" in pets:
	# start block (indentation)
	print("you have a dog.")
	# end block
elif "fish" in pets: # evaluates to True (condition met)
  print("you have a fish") # executed
else:
  print("not sure!") # executed if no conditions above is satisfied
```

### loops
#### while loops
- repeatedly execute a code block as long as a condition evaluates to `True`.
- condition of a `while` loop is always checked first before the block of code runs.

```python
i = 1
while i < 6:
  print(i)
  i = i + 1
```

#### for loops
help iterate over iterables e.g. string, list tuple etc. <br>
should not iterate and modify iterable at the same time, instead create new iterable to store the results get from changing

```python
words = ('cat', 'window', 'defenestrate')
for w in words:
  print(w, len(w))
```

### continue
continues with the next iteration of the loop

```python
for w in words:
  if len(w) < 5:
    continue
  print(w, len(w))
```

### break
breaks out of the innermost loop

```python
i = 1
flag = False
while not flag:
  while True:
    if i == 6:
      flag = True
      break
    print(i)
    i = i + 1
```

### try except finally
- a `try` and `except` block can be used to handle error.
- code which may raise an error is written in the `try` block.
- if the code raises an error, the rest of the `try` block will cease executing and the `except` code block will execute.
- code in `finally` block will be executed when either `try` or `except` block has finished execution

```python
try:
    raise Exception # throw exception raise exception_name
except:
    print("error")
finally:
    print("finally")
```

### comprehension

```python
print(list(i**2 for i in [1,2,3])) # list comprehension
print(tuple(i**2 for i in (1,2,3)))  # tuple comprehension
print(''.join(str(i**2) for i in [1,2,3])) # string from list of int
print(dict((k+1,v**2) for k,v in enumerate([1,2,3,4]))) # dict comprehension
```
