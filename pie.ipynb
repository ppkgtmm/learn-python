### default arguments

- if default args is mutable `( mutation != reassignment)` and we don't want to share them accross calls use `None` as default value and later initialize the default arg in function

### generator
- tool for creating iterators, written like regular functions but use the `yield` statement to return data
- generator function local variables and execution state are automatically saved between calls
- when generators terminate, they automatically raise `StopIteration`

````python
def reverse(data):
    for index in range(len(data)-1, -1, -1):
        yield data[index]
````

### inheritance
```python
class DerivedClassName(BaseClass1Name, BaseClass2Name, ...):
  # start block
  <statement-1>
   .
   .
  <statement-N>
  #end block
```
* when the class object is constructed, the base class is remembered and later used for resolving attribute references.
* if a requested attribute is not found in the class, the search proceeds to look in the base class.
* searching is done recursively if the base class itself is derived from some other class.
* python support multiple inheritance. if an attribute is not found in DerivedClassName, it is searched for in BaseClass1Name, then (recursively) in the base classes of BaseClass1Name, and if it was not found there, it was searched for in BaseClass2Name, and so on.
* base classes are kept from being accessed more than once

** base class can be imported from other files / folders

````python
class A:
  num = 45
  def hello(self):
    print('hello world A')

class B(A):
  # overrides method in base class (class A)
  def hello(self):
    A.hello(self) # calls method of super class
    print('hello world B')

class C(B):
  # overrides method in base class (class B)
  def hello(self):
    print('hello world C')
    print('its no.', self.num)

class D():
  num = 47
  def hello(self):
    print('hello world D')

class E(D, B):
  # overrides method in base classes
  def hello(self):
    print('hello world E')
    print('its no.', self.num) # use num from base class (class D)
````

````python
C().hello() # num of class A is printed
print()
B().hello()
print()
E().hello() # num of class D is printed
````

````python
# boolean has int as base class
print(isinstance(True, int)) # check if object (True) is instance of class (int)
print(issubclass(C, A)) # check if C (subclass) is derived from A (superclass or base class)
````

- need double _ in front of variable name to make it private to a class

`__attr_name = value`

````python
class H1:
    def __init__(self):
        # ensure that we call hello func of this class though we instantiate sub class
        self.__hello() # need to use self.__hello as hello func's 1st arg is self

    def hello(self):
      print('hello H1')

    # save copy of original hello func as private variable (only accessible inside class)
    __hello = hello

class H2(H1):

    def hello(self):
      print('hello H2')

class H3:
    '''
    self's hello func will be called
    
    when sub class is instantiated, sub class (self) hello func will 
    be called instead
    '''
    def __init__(self):
        self.hello()

    def hello(self):
      print('hello H3')

class H4(H3):

    def hello(self):
      print('hello H4')
````

````python
H1().__hello() # __hello is a private attribute
````

````python
H4().hello()
print('----------')
H2().hello()
````

### try except finally
- first, the try clause is executed.
- if no exception occurs, the except clause is skipped
- if an exception occurs during execution of try clause, the rest of the try clause is skipped and except clause is executed
- code in `finally` block (optional -- for clean up) will be executed when either `try` or `except` block has finished execution

````python
try:
    raise NameError('an error occurred') # throw exception raise exception_name(arguments)
except Exception as e: # Exception is super class of NameError; thus, NameError is caught 
    print(type(e), 'args:', e.args) # use exception stored in variable e
````

````python
try:
    raise ValueError # unhandled exception, execution will stop
except NameError as e: # NameError is neither same as nor super class of ValueError
    print('error occurred')
````

````python
try:
  raise TypeError
  # only one handler is executed by checking from top to down
except NameError as e:
  print('name error occurred')
  # executed as Exception is superclass of TypeError and comes before TypeError handler
except Exception as e:
  print('error occurred')
except TypeError as e:
  print('type error occurred')
````

````python
def raise_e():
  try:
    raise TypeError # unhandled within raise_e func
  # catch some of exception types not including TypeError
  except (NameError, ValueError, OSError) as e:
    print('error occurred')
    raise # raise the caught exception if any

try:
  raise()
except TypeError: # exception raised from raise_e func is handled here
  print('error')
finally:
  print('finally') # optional (usually for clean up)
````

### packages
python package structure should be like below. each package or sub-package should contain `__init__.py` file for initialization
<img src="https://raw.githubusercontent.com/ppkgtmm/pie/main/Package-Module-Structure.png" width=500>

when using python packages, we have to import the packages first. importing package doesn't mean executing code in package. we can import in 3 ways

1. import everything exported from package's `__init__.py`, can use functions or classes from package directly ** if they is exported

````python
from numpy import * 
print(absolute(-99), cumsum([4,5]))
````

2. import package and optionally give alias to package `import package_name as alias`, need to access function or class in package using package alias if any

````python
import numpy as np # more convenient with alias
print(np.absolute(-2), np.cumsum([4,5]))
````

````python
import numpy
print(numpy.absolute(-2), numpy.cumsum([4,5]))
````

3. import only needed functions or classes e.g.

`from module.sub1.sub2... import name as alias, ...`
can call functions, classes by their name directly




````python
from numpy import absolute as abso, cumsum, average as avg
print(abso(-2), cumsum([4,5]), avg([2,9]))
````

### relative import
- to import somthing from current directory use
`from . import name`
- to import somthing from folder in current directory use
`from .folder_name.sub_folder import name`
- add more `.` in the beginning of path to go to parent of current directory,
`from ... import name` go to parent of parent
